```python
df = pd.read_csv("../data/ethiopia.csv")
df.head()
```

```python
import numpy as np

df = df.replace(-999, np.nan)

df.isna().sum()
```

```python
df["date"] = pd.to_datetime(df["YEAR"] * 1000 + df["DOY"], format="%Y%j")

df[["YEAR", "DOY", "date"]].head()
```

```python
df["month"] = df["date"].dt.month

df[["date", "month"]].head()
```

```python
import matplotlib.pyplot as plt

df.set_index("date")["T2M"].plot(figsize=(12,5))
plt.title("Daily Mean Temperature (Ethiopia)")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
import matplotlib.pyplot as plt

df.set_index("date")["PRECTOTCORR"].plot(figsize=(12,5))
plt.title("Daily Precipitation (Ethiopia)")
plt.xlabel("Date")
plt.ylabel("Rainfall (mm/day)")
plt.show()
```

```python
import matplotlib.pyplot as plt

df.set_index("date")["PRECTOTCORR"].plot(figsize=(12,5))
plt.title("Daily Precipitation (Ethiopia)")
plt.xlabel("Date")
plt.ylabel("Rainfall (mm/day)")
plt.show()
```

```python
monthly_rain = df.groupby("month")["PRECTOTCORR"].sum()

monthly_rain.plot(kind="bar", figsize=(10,5))
plt.title("Total Monthly Precipitation (Ethiopia)")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.show()
```

```python
monthly_rain = df.groupby("month")["PRECTOTCORR"].sum()

monthly_rain.plot(kind="bar", figsize=(10,5))
plt.title("Total Monthly Precipitation (Ethiopia)")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.show()
```

```python
from scipy.stats import zscore

cols = ["T2M", "T2M_MAX", "T2M_MIN", "PRECTOTCORR", "RH2M", "WS2M", "WS2M_MAX"]

z_scores = df[cols].apply(zscore)

outliers = (z_scores.abs() > 3).sum()

outliers
```

Temperature Insight

Ethiopia shows clear seasonal variation in temperature.

Higher temperatures are observed during warmer months, while cooler months show lower T2M values.
This indicates a predictable climate cycle useful for agriculture planning.

🌧 Rainfall Insight

Rainfall is highly seasonal, concentrated in specific months.
Several months show near-zero precipitation, indicating dry periods.
This pattern suggests vulnerability to drought conditions.

⚠️ Climate Extremes

Outlier analysis shows extreme values in precipitation and temperature variables.

These extremes may represent heatwaves or heavy rainfall events.
Such events are important for flood and drought risk assessment.

🌍 COP32 Relevance

Ethiopia’s climate shows strong seasonal dependency and periodic extremes.

These insights support the need for adaptation strategies such as water management and drought preparedness.
